```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
data = loadmat('testing2agent.mat')
```

```python
y1 = data['agent1'].flatten()
y2 = data['agent2'].flatten()
```

```python
pr_list = data['price']
load_list = data['load']

ag1_load_list = load_list[:, 0]
ag2_load_list = load_list[:, 1]
ag1_pr_list = pr_list[:, 0]
ag2_pr_list = pr_list[:, 1]
```

```python
ag1_pr_list = [el for idx, el in enumerate(ag1_pr_list) if idx % 500 == 0]
ag1_load_list = [el for idx, el in enumerate(ag1_load_list) if idx % 500 == 0]
ag2_pr_list = [el for idx, el in enumerate(ag2_pr_list) if idx % 500 == 0]
ag2_load_list = [el for idx, el in enumerate(ag2_load_list) if idx % 500 == 0]
```

```python
ag1_load_list = ag1_load_list[1:]
ag2_load_list = ag2_load_list[1:]
ag1_pr_list = ag1_pr_list[1:]
ag2_pr_list = ag2_pr_list[1:]
y1 = y1[1:]
y2 = y2[1:]
```

```python
fig, ax1 = plt.subplots()

line1 = ax1.plot(np.array(ag2_pr_list)/2e3, label='Price', color='blue')
ax1.set_ylabel('Price  ($/kW)', fontsize=16, fontweight='bold')  # Increase font size and set font weight to bold

ax2 = ax1.twinx()

line2 = ax2.plot(np.array(ag2_load_list)*8, label='Demand', color='green')
ax2.set_ylabel('Demand (kW)', fontsize=16, fontweight='bold')  # Increase font size and set font weight to bold

# Combining the legend handles from both plots
lines = line1 + line2
labels = [l.get_label() for l in lines]

# Creating a single legend for both plots
legend = ax1.legend(lines, labels, loc='lower right', fontsize=16)  # Increase font size

# Set font weight to bold for legend text
for text in legend.get_texts():
    text.set_fontweight('bold')

plt.title('Agent 2: Price-Demand (PD) Curve', fontsize=14, fontweight='bold')  # Increase font size and set font weight to bold

plt.show()
```
