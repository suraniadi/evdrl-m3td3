```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
data = loadmat('testing2agent.mat')
y1 = data['agent1'].flatten()
y2 = data['agent2'].flatten()

x = np.arange(1, len(y1) + 1)  # X-axis values

plt.plot(x, y1, color='blue', label='Robust')
plt.plot(x, y2, color='green', label='Non-Robust')
plt.xlabel('Iterations', fontsize=16, fontweight='bold')
plt.ylabel('Reward', fontsize=16, fontweight='bold')
plt.title('Test Reward', fontsize=16, fontweight='bold')
plt.legend(fontsize=16, loc='best')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Make legend labels bold
legend = plt.legend(fontsize=16, loc='best')
for label in legend.get_texts():
    label.set_fontweight('bold')

plt.show()
```
